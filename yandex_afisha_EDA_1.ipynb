### 2.3 Проверка явных и неявных дубликатов

```python
# проверим неявные дубликаты, смотрим по всем столбцам, кроме 'order_id'
duplicates = final_tickets_merged_df[final_tickets_merged_df.duplicated(subset=final_tickets_merged_df.columns.difference(['order_id']), keep=False)]
display(Markdown(f"**Количество дублированных строк: {len(duplicates)}**"))
display(duplicates.head(10))
```

Как видим, неявные дубликаты представлены одинаковыми заказами, но с разными идентификаторами. Одной из причин, может являться техническая ошибка при загрузке данных. Учитывая количество неявных дубликатов эти строки можно удалить без потери качества исходных данных.

```python
# Удаляем дубликаты по всем столбцам кроме order_id
clean_final_tickets_merged_df = final_tickets_merged_df.drop_duplicates(
    subset=final_tickets_merged_df.columns.difference(['order_id']),
    keep='first'  # сохраняет первую найденную уникальную строку
).copy() # для независимости данных от исходного датафрейма

print(f"Было строк: {len(final_tickets_merged_df)}")
print(f"Стало строк: {len(clean_final_tickets_merged_df)}")
print(f"Удалено дубликатов: {len(final_tickets_merged_df) - len(clean_final_tickets_merged_df)} или {(len(final_tickets_merged_df) - len(clean_final_tickets_merged_df)) / len(final_tickets_merged_df):.2%}")
```

### 2.4 Создание нескольких новых столбцов

Для дальнейшего исследовательского анализа данных, в нашем датафрейме `clean_final_tickets_merged_df` создадим 4 новых столбца:
- **revenue_rub** — приведем выручку с заказа к единой валюте — российскому рублю;
- **one_ticket_revenue_rub** — рассчитайем выручку с продажи одного билета на мероприятие;
- **month** — выделим месяц оформления заказа;
- **season** — столбец с информацией о сезонности, будет включать такие категории, как: 'лето', 'осень', 'зима', 'весна'.

```python
# создаем новый столбц "revenue_rub"
def convert_to_rub(row):
    '''
Функция конвертирует выручку из валюты KZT в RUB с учетом исторического курса обмена на дату заказа.
    '''
    if row['currency_code'] == 'rub':
        return row['revenue']
    elif row['currency_code'] == 'kzt':
        # Находим курс на дату заказа
        rate_date = row['created_dt_msk'].date()
        rate = final_tickets_tenge_df[final_tickets_tenge_df['data'] == str(rate_date)]['curs'].values
        if len(rate) > 0:
            return row['revenue'] * rate[0] / 100  # Конвертируем тенге в рубли
        else:
            return np.nan  # Если курс не найден
    else:
        return np.nan
```

```python
# применяем функцию
clean_final_tickets_merged_df['revenue_rub'] = clean_final_tickets_merged_df.apply(convert_to_rub, axis=1)

# проверяем результат 
clean_final_tickets_merged_df.sample(5)
```

```python
# создаем новый столбц "one_ticket_revenue_rub"
clean_final_tickets_merged_df['one_ticket_revenue_rub'] = clean_final_tickets_merged_df['revenue_rub'] / clean_final_tickets_merged_df['tickets_count']

# проверяем результат
clean_final_tickets_merged_df.sample(5)
```

```python
# создаем новый столбц "month"
clean_final_tickets_merged_df['month'] = clean_final_tickets_merged_df['created_dt_msk'].dt.month

# проверяем результат
clean_final_tickets_merged_df.sample(5)
```

```python
def check_season(month):
    '''
    Функция проверяет значения в столбце 'month' и возвращает значение: 'лето', 'осень', 'зима', 'весна'.
    Если month не является числом от 1 до 12 или NaN, возвращает False.
    '''
    if pd.isna(month):  # если значение NaN
        return False
    elif month in [12, 1, 2]:  # зимние месяцы
        return 'зима'
    elif month in [3, 4, 5]:  # весенние месяцы
        return 'весна'
    elif month in [6, 7, 8]:  # летние месяцы
        return 'лето'
    elif month in [9, 10, 11]:  # осенние месяцы
        return 'осень'
    else:  # если month не в диапазоне 1-12
        return False
```

```python
# применяем функцию к датасету, создаем новый столбц "season"
clean_final_tickets_merged_df['season'] = clean_final_tickets_merged_df['month'].apply(check_season)

# проверяем результат
clean_final_tickets_merged_df.sample(5)
```

Перед проверкой нашей второй гипотезы:  
**Среднее время между заказами пользователей мобильных приложений выше по сравнению с пользователями стационарных устройств.**  
  
Сформируем нулевую и альтернативную гипотезы:  
- H<sub>0</sub> (нулевая гипотеза) - среднее время между заказами пользователей **не выше** у пользователей мобильных устройств по сравнению с пользователями стационарных устройств;
- H<sub>1</sub> (альтернативная гипотеза) - среднее время между заказами пользователей **выше** у пользователей мобильных устройств по сравнению с пользователями стационарных устройств;

Для проверки гипотезы мы выберем тест Манна-Уитни ввиду его устойчивости к ненормальному распределению выборок и содержанию выбросов, а так же к неравномерному размеру выборок. 

```python
# проверка гипотезы
stat, p_value = mannwhitneyu(mobile_user_days, desktop_user_days, alternative='greater')

# Выводим p-value и интерпретируем результат
print(f"P-value: {p_value:.4f}")
if p_value < 0.05:
    print("P-value < 0.05: Есть статистически значимые основания отвергнуть H₀ на уровне значимости 5%.")
    print("Среднее время между заказами пользователей выше у пользователей мобильных устройств по сравнению с пользователями стационарных устройств")
else:
    print("P-value >= 0.05: Нет статистически значимых оснований отвергнуть H₀ на уровне значимости 5%.")
    print("Среднее время между заказами пользователей не выше у пользователей мобильных устройств по сравнению с пользователями стационарных устройств")
```
